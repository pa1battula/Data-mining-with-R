## Question 2

```r
my2020DF <- subset(myDF, TRANSACTION_YR == "2020")
```

```r
my2020DF$TRANSACTION_MNT <- month(my2020DF$newdates)
```

```r
tapply(my2020DF$TRANSACTION_AMT,my2020DF$TRANSACTION_MNT, sum)
```

```r
total <- tapply(my2020DF$TRANSACTION_AMT,my2020DF$TRANSACTION_MNT, sum)
```

```r
my2020DF$TOTAL_AMT_BY_MONTH <- total[match(my2020DF$TRANSACTION_MNT, names(total))]
```

```r
plot(my2020DF$TRANSACTION_MNT, my2020DF$TOTAL_AMT_BY_MONTH, main= "Total Amounts by Months", xlab= "months", ylab= "Total Amounts")
```
