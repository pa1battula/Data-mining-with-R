```r
library(data.table)

products <- fread("/anvil/projects/tdm/data/icecream/combined/products.csv")
reviews <- fread("/anvil/projects/tdm/data/icecream/combined/reviews.csv")
```

```r
unique(trimws(unlist(strsplit(products$ingredients[1],",|\\(|\\)"))))
```

```r
getingredients <- function(x) {
    unique(trimws(unlist(strsplit(x,",|\\(|\\)"))))
}
```

```r
getingredients(products$ingredients[1])
```

```r
ingredient_freq <- tail(sort(table(unlist(sapply(products$ingredients, getingredients), use.names=FALSE))), n=10)
```

```r
barplot(ingredient_freq, main="Top 10 Ingredients", xlab="Ingredient", ylab="Frequency", col="pink")
```

```r
plot(tail(sort(table(unlist(sapply(reviews$text, getingredients), use.names=FALSE))), n=10))
```

By looking at the plot, you can observe which ingredients are the most commonly mentioned or used in the reviews. The ingredients with higher bars or points on the chart will be the most frequent ones, and those with lower bars or points will be less common.

This information can be valuable for understanding the preferences or common themes in the reviews. For example, if the most frequent ingredient is "chocolate," it might indicate that chocolate-related products or dishes are frequently mentioned in the reviews.

```r
products_reviews_by_rating <- function(products_df, reviews_df, myrating) {

  merge_results <- merge(products_df, reviews_df, by="key")

  
  products_reviews_results <- merge_results[merge_results$rating >= myrating, ]


  return(products_reviews_results)
}
```

```r

count_products_with_ingredient <- function(ingredient) {

  
  matching_products <- products[grepl(ingredient, products$ingredients), ]
  
  count <- nrow(matching_products)
  

  return(count)
}
```
