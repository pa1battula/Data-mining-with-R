## Question 2

```r
first_digit <- function(number) {
    number <- abs(number)
    if (number != 0) {
    while (number >= 10) {
       number <- number/10
    }
    while (number <= 1) {
     number <- number * 10
    }
    number <- floor(number)
     }   
    return(number)
}
```

```r
first_digit(-8563.02214)
```

the output is 8.

```r
first_digit(563.02214)
```

the output is 5.
